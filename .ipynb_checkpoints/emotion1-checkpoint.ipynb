```python
import pandas as pd
from scipy.stats import levene, f_oneway

# Создаем пример DataFrame
data = {
    'group': ['popular science film'] * 10 + ['lyrical sketch'] * 10 + ['crime report'] * 10,
    'emotion': [34, 35, 32, 34, 33, 34, 34, 33, 34, 37, 36, 35, 36, 38, 37, 37, 34, 36, 37, 34, 34, 32, 31, 34, 33, 32, 32, 31, 30, 31]
}
df = pd.DataFrame(data)

# Группируем данные по категориям
group1 = df['emotion'][df['group'] == 'popular science film']
group2 = df['emotion'][df['group'] == 'lyrical sketch']
group3 = df['emotion'][df['group'] == 'crime report']

# Тест Левена на однородность дисперсий
levene_stat, levene_p = levene(group1, group2, group3)
print(f"Levene's Test:\nStatistic: {levene_stat}\np-value: {levene_p}\n")

# ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)
print(f"ANOVA:\nF-statistic: {f_stat}\np-value: {p_value}")
```

```python
import matplotlib.pyplot as plt

# Данные
groups = ['popular science film', 'lyrical sketch', 'crime report']
means = [34.4, 36.1, 32.2]

# Построение графика
plt.figure(figsize=(10,6))
plt.bar(groups, means, color=['blue', 'green', 'red'])
plt.plot(groups, means, color='black', marker='o')
plt.ylim(30, 38)
plt.title('Средние значения эмоций по группам')
plt.ylabel('Среднее значение')
plt.xlabel('Группы')
plt.grid(axis='y')

# Отображение графика
plt.show()
```
